```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
#define the vocabulary 
vocab_size=10000
```

```python
### One Hot Representation 
one_hot_repr=[one_hot(words,vocab_size) for words in sent]
one_hot_repr
```

```python
#word embedding  Representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_len=8
```

```python
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_len)
# embedded_docs=pad_sequences(one_hot_repr,padding='post',maxlen=sent_len)
print(embedded_docs)
```

```python
#feature representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(vocab_size,dim,input_length=sent_len))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
